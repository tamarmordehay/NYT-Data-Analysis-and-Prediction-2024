```python
import requests
import json
import pandas as pd
import urllib.parse
import pprint
```

```python
url="https://api.nytimes.com/svc/archive/v1/2019/1.json?api-key=-"
response = requests.get(url).json()
```

```python
#pprint.pprint(response)
#see the respons in json structure
```

```python
abstract_list=[]
headline_list=[]
pub_date_list=[]
original_list=[]
person_rank=[]
news_desk_list=[]
all_keyword_list = []
lead_paragraph_list=[]
section_name_list=[]
type_of_material_list=[]

for doc in response["response"]["docs"]:
    abstract = doc.get("abstract")
    if abstract:
        abstract_list.append(abstract)   
    else:
        abstract_list.append(None)
        
    headline = doc.get("headline", {})
    main_headline = headline.get("main")
    if main_headline:
        headline_list.append(main_headline)
    else:
        headline_list.append(None)
        
    pub_date = doc.get("pub_date")
    if pub_date:
        pub_date_list.append(pub_date)
    else:
        pub_date_list.append(None)
        
    original = doc.get("byline", {}).get("original")
    if original:
        original_list.append(original)
    else:
        original_list.append(None)
        
    news_desk = doc.get("news_desk")
    if news_desk:
        news_desk_list.append(news_desk)  
    else:
        news_desk_list.append(None)    

    doc_values = []
    for keyword in doc.get("keywords", []):
        value = keyword.get("value")
        if value:
            doc_values.append(value)
        else:
            doc_values.append(None)
    all_keyword_list.append(doc_values)

    persons = doc.get("byline", {}).get("person", [])
    for person in persons:
        rank = person.get("rank")
        if rank:
            person_rank.append(rank)
        else:
            person_rank.append(None)
            
    lead_paragraph = doc.get("lead_paragraph")
    if lead_paragraph:
        lead_paragraph_list.append(lead_paragraph)
    else:
        lead_paragraph_list.append(None)  
    
    section_name = doc.get("section_name")
    if section_name:
        section_name_list.append(section_name)
    else:
        section_name_list.append(None)    
    
    type_of_material = doc.get("type_of_material")
    if type_of_material:
        type_of_material_list.append(type_of_material)
    else:
        type_of_material_list.append(None) 
```

```python
df = pd.DataFrame(list(zip(headline_list,abstract_list,pub_date_list,original_list,person_rank,all_keyword_list,lead_paragraph_list,news_desk_list,section_name_list,type_of_material_list)),columns =["headline","abstract","pub date","original","person rank","keywords","lead paragraph","news desk","section name","type of material"])
df
```

```python
df.to_csv('Data.csv', encoding='utf-8-sig', index=False)
```
